```python
from matplotlib import pyplot as plt
```

```python
plt.plot([0,1,2,3,4],[100,2,2,3,4]) #plotting 
plt.show() # showing the charts
```

# Title and labels

```python
plt.plot([0,1,2,3,4],[100,2,2,3,4])

plt.title("Random data")

plt.ylabel("Y axis")
plt.xlabel("X axis")

plt.show()
```

# Plotting Multiple lines

```python
x1 = [1,2,3,4,5,6]
y1 = [2,3,5,12,1,6]

x2 = [1,2,3,4,5,6]
y2 = [1,3,31,14,12,6]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()
```

# Adding styles

```python
from matplotlib import style
```

```python
style.use("ggplot")

x1 = [1,2,3,4,5,6]
y1 = [2,3,5,12,1,6]

x2 = [1,2,3,4,5,6]
y2 = [1,3,31,14,12,6]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()
```

# Adding legends for lines(labels)

```python
x1 = [1,2,3,4,5,6]
y1 = [2,3,5,12,1,6]

x2 = [1,2,3,4,5,6]
y2 = [1,3,31,14,12,6]

plt.plot(x1,y1 , label = "first")
plt.plot(x2,y2 , label = "second")

plt.legend() # after calling legend only labels will be shown.

plt.show()
```

# Line width

```python
x1 = [1,2,3,4,5,6]
y1 = [2,3,5,12,1,6]

x2 = [1,2,3,4,5,6]
y2 = [1,3,31,14,12,6]

plt.plot(x1,y1 , label = "first" , linewidth=12)
plt.plot(x2,y2 , label = "second")

plt.legend() 

plt.show()
```
