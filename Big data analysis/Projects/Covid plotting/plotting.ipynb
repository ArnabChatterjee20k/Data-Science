```python
numpy as np 
from matplotlib import pyplot as plt
from extractor import Extractor

extract = Extractor(state = "West Bengal")
data = extract.get_active_cases()
```

```python
x = np.array(list(data.keys()))
y = np.array(list(data.values()))
```

```python
# plot settings
# plotting a line plot after changing it's width and height
fig = plt.figure()
fig.set_figwidth(60)
fig.set_figheight(30)
  
plt.rcParams['font.size'] = 14

#plotting graphs
plt.plot(x , y)

plt.ylabel("Y axis")
plt.xlabel("X axis")

plt.title(f"District wise Covid Data {extract.state}")

plt.show()
```
